```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from __future__ import print_function, division

from sympy import *
from sympy.utilities.lambdify import lambdify  

from sympy import init_printing
init_printing() 
```

```python
# Complex conjugate in sympy
conjugate(1 + 2*I)
```

```python
# Complex Integrals

x = symbols('x', real=True)
n = symbols('n', integer=True, positive=True)
k = symbols('k', real = True, positive = True)

complex_fn = exp(-I*k*x) 

f = x**2 

result = integrate(f*complex_fn, (x, -1, 1))
result
```

```python
# plotting multiple curves on a single figure
f = sin(x)
g = cos(x)
curves = plot(f, g, (x, -2*pi, 2*pi), ylim = [-1.1, 1.1], show=False, lw = 3)
curves[0].line_color = 'firebrick'  
curves[1].line_color = 'royalblue'
curves.show()
```

```python
# Or you can specify RGB colors
# (red, green, blue)
curves[0].line_color = (0.9,0.8,1.0)
curves[1].line_color = (1.0,0.1,0.25)
curves.show()
```

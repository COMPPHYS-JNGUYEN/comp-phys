```python
%matplotlib inline

import numpy as np
from scipy.constants import pi
import matplotlib.pyplot as plt

def num_slope(y0):
    e = 1e-10
    return (np.tanh(y0+e) - np.tanh(y0-e))/(2*e)

y = np.linspace(-10, 10, 100)
y1 = np.linspace(-5., 5., 100)
an_slope = 1./np.cosh(y1)**2
numer_slope = num_slope(y1)

plt.figure()
plt.plot(y, np.sinh(y))
plt.plot(y, np.cosh(y))

plt.figure()
plt.plot(y, np.tanh(y), 'r')
plt.plot(y1, numer_slope, 'g.')
plt.plot(y1, an_slope, 'b')

plt.show()
```
